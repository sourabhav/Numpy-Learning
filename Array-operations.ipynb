**ADVANCE ARRAY MANIPULATION**

**Stacking array**

```python
a= np.array([1,2,3,4])
b= np.array([5,6,7,8])
```

```python
np.vstack((a,b))    #vertile stack
```

```python
np.hstack((a,b))     #horizontal stack (2arrays join)
```

```python
np.column_stack((a,b))
```

**Spliting array**

```python
np.hsplit(c,2)
```

```python
np.hsplit(c,4)
```

```python
np.vsplit(c,2)
```

```python
d= np.vsplit(c,4)
```
