**NUMPY INDEXING AND SLICING OF MATRIX**

```python
arr=np.arange(1,31)
arr
```

```python
arr=np.arange(1,31).reshape(6,5)
arr
```

```python
arr[5]
```

```python
arr[0,0]           #specified element
```

```python
int(arr[2,2])       #specified element
```

```python
slice=arr[0:2,1:3]
slice
```

```python
slice= arr[3:,3:]
slice
```

```python
arr[:,2]
```

**BOOLEAND INDEXING**

```python
arr=np.arange(11,21)
arr
```

```python
bool_index= arr%2==0     #even numbers
bool_index  
```

```python
arr= arr[bool_index]
arr
```
